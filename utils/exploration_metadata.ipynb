```python
metadata_df_1 = metadata_df_1[metadata_df_1['contenttype'] == 'Documentation']
```

```python
# Keep only the rows that have 'content_type' as 'document'

metadata_df_2 = metadata_df_2[metadata_df_2['content_type'] == 'Documentation']
metadata_df_3 = metadata_df_3[metadata_df_3['content_type'] == 'Documentation']
metadata_df_4 = metadata_df_4[metadata_df_4['content_type'] == 'Documentation']
```

```python
metadata_columns = [
    'submitted_for_country',
    'helpdeskname',
    'segmentname',
    'content_type',
    'content_subtype',
]

rename_map = {
    'submittedforcountry': 'submitted_for_country',
    'contenttype': 'content_type',
    'contentsubtype': 'content_subtype',
}


source_dataframes = [metadata_df_1, metadata_df_2, metadata_df_3, metadata_df_4]

normalized_sources = []

for df in source_dataframes:
    tmp = df.rename(columns=rename_map).copy()
    tmp = tmp[tmp['content_type'].eq('Documentation')]
    tmp = tmp[['contentid', *metadata_columns]]
    normalized_sources.append(tmp)

all_metadata_df = (
    pd.concat(normalized_sources, ignore_index=True)
      .drop_duplicates(subset=['contentid'])
)

all_metadata_df.head()
```

```python
# Delete rows that have a null value in the 'content_type' column
all_metadata_df = all_metadata_df.dropna(subset=['content_type'])

```

```python
# Add the filename to each element in the all_metadata_df DataFrame by merging it with the main_metadata_df DataFrame on the 'contentid' column and the 'content_id' column, respectively. After the merge, drop the 'content_id' column from the resulting DataFrame.
all_metadata_df = all_metadata_df.merge(
    main_metadata_df[['content_id', 'file_name']],
    left_on='contentid',
    right_on='content_id',
    how='left'
).drop(columns='content_id', errors='ignore')
```

```python
# Fill the null values in the 'file_name' column with 'file_not_available_yet'
all_metadata_df['file_name'] = all_metadata_df['file_name'].fillna('file_not_available_yet')
```

```python
# Add the path to each element.

all_metadata_df["path"] = (
    "parsed_documents/" +
    all_metadata_df["file_name"].str.replace(r"\.pdf$", ".txt", regex=True)
)
```

```python
# Export the dataframe to JSON format, each row is an element of the JSON.
all_metadata_df.to_json("metadata_map.json", orient="records", indent=4)
```
